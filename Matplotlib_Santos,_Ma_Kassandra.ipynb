```python
df=pd.read_csv('Technological-Products-Sample-Data - Tech Products - Date Sorted.csv')
df
```

# ***EACH DEVICE'S DATE OF RELEASE AND SALES***

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(15, 6))
plt.plot(df['Date of Release'], df['Sales (USD)'], color='red', marker='o', label='Sales', linewidth = 3, linestyle = 'dotted')  # Add label

plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Overall Sales Trend of all of the Brands Over Time')

plt.xticks(rotation=45)

plt.legend(loc = 'lower right')

plt.show()
```

# ***USING MULTILINE PLOT TO READ ALL PRODUCTS***

```python
import matplotlib.pyplot as plt
import pandas as pd


fig, ax = plt.subplots(figsize=(30, 10))


ax.plot(df[df['Brand'] == 'Acer']['Date of Release'], df[df['Brand'] == 'Acer']['Sales (USD)'], label='Acer')
ax.plot(df[df['Brand'] == 'Apple']['Date of Release'], df[df['Brand'] == 'Apple']['Sales (USD)'], label='Apple')
ax.plot(df[df['Brand'] == 'Asus']['Date of Release'], df[df['Brand'] == 'Asus']['Sales (USD)'], label='Asus')
ax.plot(df[df['Brand'] == 'Dell']['Date of Release'], df[df['Brand'] == 'Dell']['Sales (USD)'], label='Dell')
ax.plot(df[df['Brand'] == 'Google']['Date of Release'], df[df['Brand'] == 'Google']['Sales (USD)'], label='Google')
ax.plot(df[df['Brand'] == 'HP']['Date of Release'], df[df['Brand'] == 'HP']['Sales (USD)'], label='HP')
ax.plot(df[df['Brand'] == 'Lenovo']['Date of Release'], df[df['Brand'] == 'Lenovo']['Sales (USD)'], label='Lenovo')
ax.plot(df[df['Brand'] == 'Microsoft']['Date of Release'], df[df['Brand'] == 'Microsoft']['Sales (USD)'], label='Microsoft')
ax.plot(df[df['Brand'] == 'Samsung']['Date of Release'], df[df['Brand'] == 'Samsung']['Sales (USD)'], label='Samsung')
ax.plot(df[df['Brand'] == 'Sony']['Date of Release'], df[df['Brand'] == 'Sony']['Sales (USD)'], label='Sony')

ax.set_xlabel('Year')
ax.set_ylabel('Sales')
ax.set_title('Sales Trend for Each Brand')


plt.legend(loc='lower right')
plt.xticks(rotation=45)


plt.show()
```

# ***HISTOGRAM OF THE MOST COMMON COUNTRY OF ORIGIN***

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['Country of Origin'], color='violet',bins=6, edgecolor='black')
plt.title('Histogram of Country of Origin for All Devices')
plt.xlabel('Name of the country')
plt.ylabel('Number of Devices produced by a country')
plt.show()
```

# ***PIE CHART THAT SHOWS EACH BRAND COUNTRY OF ORIGIN***

```python
import matplotlib.pyplot as plt
import pandas as pd

country_counts = df['Country of Origin'].value_counts()


plt.figure(figsize=(12, 10))
plt.pie(country_counts.values, labels=country_counts.index, autopct='%1.1f%%', startangle=90)
plt.legend()
plt.legend(loc = 'lower left')

plt.title('Distribution of All Device by Country of Origin')
plt.axis('equal')
plt.show()
```
