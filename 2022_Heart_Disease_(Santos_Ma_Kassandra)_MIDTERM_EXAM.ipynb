# ***QUESTION***

1. What can you infer about the overall general health of all of the participants

    The histogram shows that the highest frequency would be "very good" and "good" suggesting that most of the participants are in a healthy condition with a few of the participants are in a not so good health condition

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
df['GeneralHealth'] = df['GeneralHealth'].astype(str)

plt.hist(df['GeneralHealth'], color='pink', bins=6, edgecolor='black')
plt.title('Histogram of General Health of the Participants')
plt.xlabel('General Health')
plt.ylabel('Number of Participants')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# ***Questions***

1.   Compare the percentage of the male participants who had a heart attack to the percentage of the female participants who had a heart attack. What can you infer about the pie chart.

      male participants have a higher percentage of having a heart attack with a percentage of 7.4% compared to the female participants who have a lower reported incident of having a heart attack with a percentage of 4.1%





```python
import pandas as pd
import matplotlib.pyplot as plt

female_df = df[df['Sex'] == 'Female']
heartattack_counts = female_df['HadHeartAttack'].value_counts()

plt.figure(figsize=(8, 7))
plt.pie(heartattack_counts, labels=heartattack_counts.index, autopct='%1.1f%%',startangle=90)
plt.title('Data analysis of the Female Participants Who Had Heart Attack')
plt.axis('equal')
plt.legend()
plt.legend(loc = 'lower right')

plt.show()


male_df = df[df['Sex'] == 'Male']
heartattack_counts = male_df['HadHeartAttack'].value_counts()

plt.figure(figsize=(8, 7))
plt.pie(heartattack_counts, labels=heartattack_counts.index, autopct='%1.1f%%',startangle=90)
plt.title('Data analysis of the Male Participants Who Had Heart Attack')
plt.axis('equal')
plt.legend()
plt.legend(loc = 'lower right')

plt.show()
```
