#***QUESTIONS FORMULATED:***

1. How does the distribution of BMI relate to the  heart disease among females in the dataset? (FIGURE 1)

  The Data set shows that majority of the female participants tend to have a higher Body Mass Index or otherwise known as BMI but do not have any heart related diseases whatsoever compared to the other female participants whom they do have heart related diseases but have a lower body mass index. This output shows an interesting information as it is expected that does who have a normal or lower BMI count would have less chance of having a heart related disease which in this case did not occur, this output would mean that other underlying causes and the individuals’ lifestyles also have a big part in their health




```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('heart_2020_cleaned.csv')

plt.figure(figsize=(10, 6))
fm_df = df[df['Sex'] == 'Female']

plt.bar(fm_df['HeartDisease'], fm_df['BMI'], color='lightblue', label='Female')

plt.xlabel('Heart Disease')
plt.ylabel('BMI')
plt.title('The Distribution of BMI in Relation to Heart Disease for Females')
plt.legend(loc='upper right')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv('heart_2020_cleaned.csv')

plt.figure(figsize=(10, 6))
male_df = df[df['Sex'] == 'Male']

plt.bar(male_df['HeartDisease'], male_df['BMI'], color='darkgreen', label='BMI')

plt.xlabel('Heart Disease')
plt.ylabel('BMI  COUNT')
plt.title('THe distribution BMI count in Relation to Heart Disease for Males')
plt.legend()

plt.show()
```
