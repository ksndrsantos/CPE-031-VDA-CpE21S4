***2. ANALYZE THE DATA AND GENERATE THE APPROPRIATE PLT INCLUDING ITS PROPERTIES***

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("Technological-Products-Sample-Data - Tech Products.csv")

plt.figure(figsize=(30, 6))
plt.plot(df['Date of Release'], df['Price (USD)'], color='hotpink', marker='*', label='Price')  # Add label

plt.xlabel('Date of Release')
plt.ylabel('Price in USD')
plt.title('Line Plot of Date of Release vs Price in USD Overtime')

plt.xticks(rotation=45)

plt.legend() # Add the legend

plt.show()
```
